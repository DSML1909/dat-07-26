This notebook will walk us through how you can stack multiple dataframes on top of each other from multiple file sources

```python
# do our imports
import pandas as pd
import os
```

```python
# we can load in the dataframes separately and stack on top of teach other one by one
df1 = pd.read_csv('data-for-concatenation/sample1.csv')
df2 = pd.read_csv('data-for-concatenation/sample2.csv')
```

```python
# our first dataframe
df1
```

```python
# our second
df2
```

```python
# pd.concat will stack the two on top of each other
pd.concat([df1, df2])
```

```python
# and we can save it like so
df3 = pd.concat([df1, df2])
```

If the data is in once location, you can also do this a bit more concisely with the help of the `os` module

```python
# we can use listdir() to get the contents of a single directly
os.listdir('data-for-concatenation')
```

```python
# since this is a list of all the csv files, we can load them add in with a comprehension
df4 = pd.concat([
    pd.read_csv(f'data-for-concatenation/{file}')
    for file in os.listdir('data-for-concatenation')
])
```
